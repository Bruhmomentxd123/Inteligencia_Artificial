```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/dilp90/InteligenciaArtificial_y_RedesNeuronales_UANL_FIME/main/MachineLearning/Datasets/robot_sensors.csv"

df = pd.read_csv(url)

df.head()
```

```python
df.info()
df.describe()
```

```python
df_clean = df.dropna()
```

```python
df_normalized = df_clean.copy()

for column in df_clean.columns:
    min_val = df_clean[column].min()
    max_val = df_clean[column].max()
    df_normalized[column] = (df_clean[column] - min_val) / (max_val - min_val)

df_normalized.head()
```

```python
plt.figure()
plt.plot(df_normalized["Tiempo (s)"], df_normalized["Distancia (cm)"])
plt.xlabel("Tiempo (normalizado)")
plt.ylabel("Distancia (normalizada)")
plt.title("Distancia vs Tiempo (Normalizado)")
plt.show()
```
